```python
#ensure all hours are presented for each day

for date in df['date'].unique():
    #print(date)
    dayofweek = df[df['date'] == date]['day_of_week'].unique()[0]
    #print(dayofweek)
    present_hours = df[df['date'] == date]['hour'].tolist()
    missing_hours = set(all_hours) - set(present_hours)
    #print(missing_hours)
    for hour in missing_hours:
        #print(hour)
        df = pd.concat([df,pd.DataFrame({'date':date,'day_of_week':dayofweek,'hour':hour, 'count': 0}, index = [0])], ignore_index = True)
```

```python
monday_df = df.loc[df['day_of_week'] == 0]
#monday_data = 
```

```python
data = {}
for h in all_hours:
    data[h] = monday_df.loc[monday_df['hour'] == h]['count'].to_list()
```

```python
data[12]
```

```python
# Assuming 'data' is a dictionary where keys are hours and values are lists of traffic values
# e.g., data = {1: [100, 120, ...], 2: [150, 145, ...], ...}
bootstrap_samples = {hour: [] for hour in data.keys()}  # Initializing a dictionary to store bootstrap samples per hour

bootstrap_samples
```

```python
N = 10000  # Number of bootstrap samples
for _ in range(N):
    for hour, hourly_data in data.items():
        resampled_data = np.random.choice(hourly_data, len(hourly_data), replace=True)
        bootstrap_samples[hour].extend(resampled_data)

# Compute desired metrics on bootstrap_samples, e.g., mean, variance, etc.
hourly_means = {hour: np.mean(samples) for hour, samples in bootstrap_samples.items()}
hourly_variances = {hour: np.var(samples) for hour, samples in bootstrap_samples.items()}

print("Hourly Means:", hourly_means)
print("Hourly Variances:", hourly_variances)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Simulate age data for 1000 individuals
ages = np.random.uniform(20, 60, 1000)


# Simulate height data: height = 150 + 0.5*age + random_error
heights = 150 + 0.5*ages + np.random.normal(0, 5, 1000)
```

Visualize the distribution of heights:

```python
sns.histplot(heights, kde=True)
plt.title("Distribution of Heights")
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.show()
```

Visualize the relationship between age and height with the regression line:

```python
sns.regplot(x= ages, y= heights, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title("Age vs Height with Regression Line")
plt.xlabel("Age")
plt.ylabel("Height")
plt.show()
```

```python
import pandas as pd

# Fit a simple linear regression
X = sm.add_constant(ages)
model = sm.OLS(heights, X).fit()
predictions = model.predict(X)
```

```python
model.summary()
```

```python
# Calculate residuals
residuals = heights - predictions
pred_mu_diff = [p - np.average(heights) for p in predictions]
```

```python
sns.histplot(pred_mu_diff, kde=True)
```

```python
sns.histplot(residuals, kde=True)
```

```python
# Create dataframe for plotting
df = pd.DataFrame({'Age': ages, 'Height': heights, 'Predicted Height': predictions, 'Residual': residuals})
```

```python
# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Height', data=df, label='Observed Heights')
sns.lineplot(x='Age', y='Predicted Height', data=df, color='red', label='Regression Line')
plt.vlines(df['Age'], df['Predicted Height'], df['Predicted Height'] + df['Residual'], colors='gray', linestyle='dashed', label='Residuals')
plt.legend()
plt.title('Decomposition of Variability in Height')
plt.show()
```
